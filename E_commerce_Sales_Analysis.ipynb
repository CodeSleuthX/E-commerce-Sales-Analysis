```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Data Loading
file_path = './drive/MyDrive/e_com_data.csv'
df = pd.read_csv(file_path)
```

```python
# Step 2: Data Cleaning
# Handling missing values
df.dropna(inplace=True)
```

```python
# Removing duplicates
df.drop_duplicates(inplace=True)
```

```python
# Step 3: Exploratory Data Analysis (EDA)
# Summary statistics
print("Summary Statistics:")
print(df.describe())
```

```python
# Visualizing distribution of sales quantities
plt.figure(figsize=(10, 6))
sns.histplot(df['Quantity'], bins=20, kde=True)
plt.title('Distribution of Sales Quantities')
plt.xlabel('Sales Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing distribution of product prices
plt.figure(figsize=(10, 6))
sns.histplot(df['UnitPrice'], bins=20, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing the relationship between variables using scatter plots
sns.pairplot(df[['Quantity', 'UnitPrice']])
plt.show()
```

```python
# Step 3: Product Analysis
# Top-selling products based on sales quantity
top_products = df.groupby('StockCode')['Quantity'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Selling Products:")
print(top_products)
```
